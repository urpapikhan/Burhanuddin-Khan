```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import missingno as msno
import seaborn as sns

df=pd.read_csv("https://raw.githubusercontent.com/nytimes/covid-19-data/master/live/us-states.csv")
df.head()
```

```
df.isnull().sum()
```

```
df.nunique()
```

```
df.describe()
```

```
sns.catplot(data=df,kind='box', height=4, aspect=2)
```

```
sns.catplot(x='deaths', data=df, kind='count', height=3, aspect=1.1)
```

```
df.hist(figsize=(15, 10), bins=20, xlabelsize=10, ylabelsize=10) 
```

```
sns.lmplot(x='cases', y='deaths', data=df)
```

```
sns.lmplot(x='cases', y='deaths', data=df,
           fit_reg=False, # No regression line
           hue='fips')
```

```
fig, ax = plt.subplots(figsize=(11, 9))
sns.heatmap(df)
plt.show()
```
