# Data Set
The data set that was chosen is data on the rate and percentage of people diagnosed with cancer based on age,location, race and sex. The subcategories consist of the type of cancer. Our data set includes breast cancer, colon cancer, and lung cancer. 

```
#Load Libraries
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import missingno as msno
import seaborn as sns

#Import Data
df = pd.read_csv("https://raw.githubusercontent.com/CDeMaio19/BCS109/master/cancer.csv")
df.head()
```

#Compiling a Bar Graph
A bar graph can be compiledfrom the data set by selecting what information to compute. In the following graph the states are listed on the Y axis and the total cancer rate is listed on the X axis. THe following bar graph is a representation of the total rate of cancer varying by state.

```

y = df['Total.Rate']

plt.figure(figsize=(15,4))
plt.bar(df.State, y)

plt.xlabel("State")
plt.ylabel("Total Rate")
plt.title("Cancer Rate Per State")
plt.xticks(rotation=90)
plt.show()
```

# Compiling a Bar Plot- Race
A Bar plot can be  constructed to represent two different categories in one diagram. THe following Bar Plot consists of the rate of cancer varying by states, and is topped off by a representation of the different races that are included in this study by state. They are color cordinated and can be changed in the code of the bar plot design. The different race color cordinations are as followed:


*   Black- Blue
*   White/Non Hispanic- Red
*   White- Green
*   Asian-Yellow






```
y1 = df['Rates.Race.White']
y2 = df['Rates.Race.White non-Hispanic']
y3 = df['Rates.Race.Black']
y4 = df['Rates.Race.Asian']

plt.figure(figsize=(15,4))



plt.bar(df.State, y3, label ="Black", color = 'b')
plt.bar(df.State, y2, label ="White non-Hispanic", color = 'r')
plt.bar(df.State, y1, label ="White", color = 'g')
plt.bar(df.State, y4, label ="Asian", color = 'y')
plt.plot()

plt.xlabel("State")
plt.ylabel("Total Rate")
plt.title("Cancer Rate Per Race")
plt.xticks(rotation=90)
plt.legend()
plt.show()
```

#Compiling Bar Plot- Age
Another Bar Plot is used to compute the age categories based on state. The following bar plot consists of The age categories of under 18 to above the age of 64. This helps break down the data set into specific need based data.THe age categories are color organized as the following:


*   Under 18- Yellow
*   18-45- Green
*   45-64- Red
*   Above 64- Blue





```
y1 = df['Rates.Age.< 18']
y2 = df['Rates.Age.18-45']
y3 = df['Rates.Age.45-64']
y4 = df['Rates.Age.> 64']

plt.figure(figsize=(15,4))



plt.bar(df.State, y4, label =">64", color = 'b')
plt.bar(df.State, y3, label ="45-64", color = 'r')
plt.bar(df.State, y2, label ="18-45", color = 'g')
plt.bar(df.State, y1, label =">18", color = 'y')
plt.plot()

plt.xlabel("State")
plt.ylabel("Total Rate")
plt.title("Cancer Rate Per Age")
plt.xticks(rotation=90)
plt.legend()
plt.show()
```

#Pie Chart
The following pie chart lists the percentage of Asians in the United states affected by cancer. The percentage is listed by state in tge pie chart itself.

```
import matplotlib.pyplot as plt
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/CDeMaio19/BCS109/master/cancer.csv")
y4 = df['Rates.Race.Asian']
colors = ["#1f77b4", "#ff7f0e", "#2ca02c", "#d62728", "#8c564b"]  
plt.pie(y4, labels=df.State, colors=colors,
autopct='%1.1f%%', shadow=True, startangle=140)
plt.title("Pie Chart of asians that have cancer by state")
fig = plt.gcf()
fig.set_size_inches(30,15)
plt.show()
```
